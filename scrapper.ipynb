```python
import requests
from bs4 import BeautifulSoup 
```

```python
URL = 'https://www.mdbg.net/chinese/dictionary?page=worddict&wdrst=0&wdqb=super'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find('div',class_='hanzi')
# table_elements = soup.findAll('table')[1].findAll('tr')
```

```python
# print(table_elements)
print(results)
```

```python
table_body=soup.find('tbody')
rows = table_body.find_all('tr')
for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    print (cols)
```
